## Distancia Euclidiana entre números SVN

```python
def euclideanNeu(a1,a2):
    a=0
    c=len(a1)    
    for i in range(c):        
        a=a+pow(a1[i][0]-a2[i][0],2)+pow(a1[i][1]-a2[i][1],2)+pow(a1[i][2]-a2[i][2],2)
    a=pow(1.0/3.0*a,0.5)
    return(a)
```

## Ejemplo de uso de la distancia Euclidiana

## Operador SVNWA

```python
def SVNWA(list,W):
    t=1
    i=1
    f=1
    c=0
    for j in list:
        t=t*(1-j[0])**W[c]
        i=i*j[1]**W[c]
        f=f*j[2]**W[c]
        c=c+1
    return (1-t,i,f)
    
```

## Ejemplo de uso SVNWA

```python
A=[MDB,B,MDB]
W = [0.55, 0.26, 0.19] # W:Vector de pesos
SVNWA(A,W)
```

## Operador SVNGA

```python
def SVNGA(list,W):
    t=1
    i=1
    f=1
    c=0
    for j in list:
        t=t*j[0]**W[c]
        i=i*j[1]**W[c]
        f=f*j[2]**W[c] 
        c=c+1
    return (t,i,f)
```

```python
A=[MDB,B,MDB]
W = [0.55, 0.26, 0.19] # W:Vector de pesos
SVNGA(A,W)
```
